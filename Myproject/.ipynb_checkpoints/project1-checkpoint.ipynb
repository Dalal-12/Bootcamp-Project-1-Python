```python
import sqlite3
from datetime import datetime, timedelta
conn = sqlite3.connect('mylibrary.db')
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS Books
                   (id INTEGER ,
                   name TEXT PRIMARY KEY,
                   available BOOLEAN,
                   borrower TEXT)''')

cursor.execute('''CREATE TABLE IF NOT EXISTS users
                   (id INTEGER ,
                   name TEXT PRIMARY KEY,
                   is_active text,
                   borrowed_books_num INTEGER,
                   borrowed_books_name TEXT,
                   date_of_borrowing TEXT) ''')
conn.commit()

def add_user():
    try:
        new_user = input("Enter the user name: ")
        cursor.execute("INSERT INTO users (name, is_active, borrowed_books_num, borrowed_books_name) VALUES (?, ?, ?, ?)", (new_user, 'Active', 0, ''))
        conn.commit()
        print(f"The user {new_user} has been added successfully.")
        cursor.execute("SELECT * FROM users ORDER BY id DESC LIMIT 1")
        record = cursor.fetchone()
        print(f"ID: {record[0]} \nName: {record[1]} \nStatus: {record[2]} \nNumber of Borrowed Books: {record[3]} \nBooks: {record[4]} ")
        main()
    except sqlite3.IntegrityError as e:
        print(f"Error while adding user: {e}")
    except sqlite3.OperationalError as e:
        print(f"Error while adding user: {e}")
        main()
        
    
def is_book_available(book_name):
    cursor.execute("SELECT available FROM Books WHERE name=?", (book_name,))
    book = cursor.fetchone()
   #book will contain the result 1 or 0
    if book :
        if book[0] == 0: 
            return True
        else:
            return False
    else:
        return False
    
def retrive_user_info(user_id):
    cursor.execute(f"SELECT * FROM users where id='{user_id}' LIMIT 1")
    user_info=cursor.fetchone()
    if user_info:
        print(f'''Name:{user_info[1]}          ID:{user_info[0]}          Status :{user_info[2]}
        \nNumber of borrowed book:{user_info[3]}          Books:{user_info[4]} ''')
    else:
        print("User not found")
        exit()
    
def start_borrow_op(user_id,book_name):
        cursor.execute("SELECT * FROM users WHERE id=?", (user_id,))
        user = cursor.fetchone()
        if user[3] > 2 :
            print("Sorry You are not Allowed to borrow books , you reached the maximum number 2")
        else:
            borrowed_books_num = user[3] + 1
        if user[4] != "":
            borrowed_books_name = f"{user[4]}, {book_name}"
        else:
            borrowed_books_name = book_name
        date_of_borrowing = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # now update the table in database
        cursor.execute("UPDATE users SET borrowed_books_num=?, borrowed_books_name=?, date_of_borrowing=?, WHERE id=?",
                      (borrowed_books_num, borrowed_books_name, date_of_borrowing, user_id))
        cursor.execute("UPDATE Books SET available=?, borrower=? WHERE name=?",
                      (1, user_id, book_name))
        conn.commit()
        print(f"{book_name} is borrowed by User ID {user_id} Successfully.")
        is_user_active(user_id)
        conn.close()


def return_book():
    is_active=""
    book_name = input("Enter the name of the book you want to return: ")
    book_name = book_name.upper()
    cursor.execute("SELECT * FROM Books WHERE name=?", (book_name,))
    book = cursor.fetchone()
    if book and book[2] == 1: #not available means really borrowed
        borrower_id = book[3] #id for the user who borrowed the book
        cursor.execute("UPDATE Books SET available=?, borrower=? WHERE name=?",(1, None, book_name))
        conn.commit()
        cursor.execute("SELECT * FROM Users WHERE id=?", (borrower_id,))
        user = cursor.fetchone()
        borrowed_books_num = user[3] + 1
        borrowed_books_name = user[4] + ", " + book_name
        cursor.execute("UPDATE Users SET borrowed_books_num=?, borrowed_books_name=?,WHERE id=?", (borrowed_books_num, borrowed_books_name,borrower_id))
        conn.commit()
    else:
        print("Wrong book! This book is not borrowed by anyone.")
      
    
def main():
    while True:
        try:
            action=int(input("what you want? Type the number of your choice\n1. Add new user\n2. Borrow a Book \n3. Return a Book\n "))
            if action ==1:
                add_user()
                break
            elif action == 2:
                user_id=int(input('enter user id :'))
                retrive_user_info(user_id)
                book_name=input("Enter the name of book you want to borrow: ")
                book_name=book_name.upper()
                is_book_available(book_name)
                if is_book_available(book_name) :
                    start_borrow_op(user_id,book_name)
                else:
                    print("This book not available for borrowing")
            elif action == 3 :
                return_book() 
                break
            else :
                print("Please Enter the correct choice number :")
        except Exception as e:
            print(f"An error occurred: {e}")

main()
```
